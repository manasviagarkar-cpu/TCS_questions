Given a string S consisting of only two types of characters: asterisks (*) and hashes (#). Your task is to evaluate the validity of the string. A string is considered valid if the total number of * characters equals the total number of # characters.

#logic
S=string , consisting of *,#

```python
S="***###"
A=S.count("*")
B=S.count("#")
def validity(A,B):
    if A==B:
        print("valid")
    else:
        print("invalid")

validity(A,B)
```
