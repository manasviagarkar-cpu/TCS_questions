#wheel question
An automobile company manufactures both Two-Wheelers (TW) and Four-Wheelers (FW). Given the total number of vehicles (V) and the total number of wheels (W), your task is to calculate the exact number of two-wheelers and four-wheelers produced

#logic
#### TW +FW = V --1
####2*TW + 4*FW =W --2


####after solving TW, FW we get
 # TW=4V-W/2
 # FW=W-2V/2 OR V-TW


```python
def find_TW_FW(V, W):
  if W<2 or W%2 != 0 or V>W:
    print("Invalid input")
  else:
    FW=((W-2*V)/2)
    TW=V-FW
    print("Number of two-wheelers:",TW)
    print("Number of four-wheelers:",FW)

#EXAMPLE
V=106
W=200
find_TW_FW(V,W)
```
